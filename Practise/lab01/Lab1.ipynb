```python
#ve do thi
x=[1,2,3,4,5]
y=[10,16,20,30,25]
x1=[1,2,3,4,5]
y1=[10,18,20,31,25]
plt.title('Dothi')
plt.xlabel('do dai')
plt.ylabel('do rong')
plt.plot(x,y,'bo-',color='red', label='Du lieu')
plt.plot(x1,y1,color='blue', marker='o',linestyle='--',label='Data')
plt.legend()
plt.show()
```

```python
# ve bieu do bar chart
categories=['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]
plt.bar(categories, values,color='r',alpha=0.6,label='A:1') #alpha la do trong suot cua color
plt.title('Bar Chart')
plt.legend()
plt.xlabel('danh muc')
plt.ylabel('Gia tri')
plt.show()
```

```python
#ve bieu do pie chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_color=['red','green','blue','yellow']
plt.pie(sizes,labels=my_labels,colors=my_color, autopct='%1.2f%%') #autopct:auto percentage
plt.title('Pie Chart')
plt.show()
```

#Scipy


```python
from scipy import linalg
#giai he phuong trinh tuyen tinh
# dinh nghia phuong trinh tuyen tinh
A= np.array([[2,1],[3,2]])
b=np.array([5,7])
x= linalg.solve(A,b)
print(x)
```

```python
from scipy import integrate # tich phan

# tinh tich phan cua mot ham so
#ding nghia ham f(x)= x^2
def my_function(x):
  return x**2
# tinh tich phan cua f(x) tu 0-1
integral=integrate.quad(my_function,0,1)
print(integral)

```

```python
#tinh gia tri rieng va vector rieng cua mot ma tran

A=np.array([[2,1],[3,2]])

evals,evecs= linalg.eig(A)
print(evals)
print(evecs)
```
