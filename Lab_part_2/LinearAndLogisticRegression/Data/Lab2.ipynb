```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('/content/wines.csv')

if 'type' in df.columns:
    df = pd.get_dummies(df, columns=['type'], drop_first=True)

y = df['quality']
X = df.drop(columns=['quality'])

X = X.astype(float)

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
titanic = pd.read_csv("/content/titanic.csv")
```

```python
df = pd.read_csv('/content/titanic.csv')

df['sex'] = df['sex'].map({0: 'Male', 1: 'Female'}).astype('category')

df['embarked'] = df['embarked'].map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}).astype('category')

pclass_order = ['1st', '2nd', '3rd']
df['pclass'] = df['pclass'].map({1: '1st', 2: '2nd', 3: '3rd'})
df['pclass'] = pd.Categorical(df['pclass'], categories=pclass_order, ordered=True)

model = smf.logit("survived ~ C(pclass) + C(sex) + age + sibsp + parch + fare + C(embarked)", data=df).fit()

print(model.summary())
```

```python
import pandas as pd
import statsmodels.api as sm
import warnings
```

```python
wines["quality_grp"] = pd.cut(
    wines["quality"],
    bins=[0, 5, 7, 10],
    labels=[0, 1, 2]
).astype(int)
```

```python
df = pd.read_csv('wines.csv')

if 'type' in df.columns:
    df = pd.get_dummies(df, columns=['type'], drop_first=True)

y = df['quality']
X = df.drop(columns=['quality'])

X = X.astype(float)

X = sm.add_constant(X)

model = sm.MNLogit(y, X).fit(method='bfgs', maxiter=1000)

try:
    print(model.summary())
except ValueError as e:
    warnings.warn(f"Could not generate full model summary due to: {e}. This often indicates issues like multicollinearity or perfect separation.")
    print("\n--- Model Coefficients (Summary Unavailable) ---")
    print("Optimization terminated successfully, but the full summary could not be generated because the covariance matrix of the parameters could not be computed (likely due to a singular Hessian).")
    print("This often happens with issues like perfect multicollinearity or perfect separation in the data, preventing the calculation of standard errors and p-values.")
    print("Estimated coefficients:")
    print(model.params)
    print("\nConsider reviewing your independent variables for multicollinearity or perfect separation.")
```
