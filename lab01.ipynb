```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
print(titanic.info())
print(titanic.head())
```

#Exercise 1.2

```python
embarked_dict = {'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}
titanic['embarked'] = titanic['embarked'].map(embarked_dict)
titanic['embarked'] = pd.Categorical(
    titanic['embarked'],
    categories=['Cherbourg', 'Queenstown', 'Southampton'],
    ordered=False
)
print(titanic['embarked'].unique())
print(titanic['embarked'].value_counts())
```

```python
sex_dict = {'female': 'Female', 'male': 'Male'}
titanic['sex'] = titanic['sex'].map(sex_dict)
titanic['sex'] = pd.Categorical(titanic['sex'], categories=['Female', 'Male'], ordered=False)
print(titanic['sex'].unique())
print(titanic['sex'].value_counts())
```

```python
survived_dict = {1: 'Yes', 0: 'No'}
titanic['survived'] = titanic['survived'].map(survived_dict)
titanic['survived'] = pd.Categorical(titanic['survived'], categories=['Yes', 'No'], ordered=False)
print(titanic['survived'].unique())
print(titanic['survived'].value_counts())
```

```python
pclass_dict = {1: '1st', 2: '2nd', 3: '3rd'}
titanic['pclass'] = titanic['pclass'].map(pclass_dict)
titanic['pclass'] = pd.Categorical(
    titanic['pclass'],
    categories=['1st', '2nd', '3rd'],
    ordered=True
)
print(titanic['pclass'].unique())
print(titanic['pclass'].value_counts())
```
